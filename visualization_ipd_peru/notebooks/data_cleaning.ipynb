```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv(r"..\data\raw\Data_DeportistasEventos.csv", encoding='latin-1', sep=';')
df.head(5)
```

```python
df['PUESTO'].unique()
```

```python
df.info()
```

```python
# Columnas no necesarias
df = df.drop(columns=['COLECTIVO','DEPARTAMENTO','PROVINCIA','DISTRITO','UBIGEO'])
```

```python
# convirtiendo columnas fecha de Int a Datetime
columns_to_convert = ['FECHA_CORTE','FECHA_INICIO', 'FECHA_FIN', 'FECHA_PUBLICACION']

for column in columns_to_convert:
    df[column] = pd.to_datetime(df[column].astype(str), format='%Y%m%d')
```

```python
df.info()
```

```python
df.describe()
```

```python
cols = ['FEDERACION','PAIS_EVENTO','CIUDAD_EVENTO','ESPECIALIDAD' ]

for col in cols:
    
    unique_values = df[col].unique()
    print(f"{col}")
    print(f"valores únicos: {unique_values}")
    print("=="*40)
```

```python
df["cant_medallas"] = df['PUESTO'].isin(['ORO', 'PLATA','BRONCE']).astype(int)
```

```python
df.groupby(by='FEDERACION', as_index=False)\
    .agg(
        total_medallas=("cant_medallas","sum")
    ).sort_values(by="total_medallas", ascending=False)
```
